```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Замена каналов RGB местами
Загрузим изображение и изменим его цветовую модель на RGB (все изображения считываются в BGR)

```python
img = cv2.imread('images/jaguar_car.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

cv2.split() разделяет цветовые каналы, а cv2.merge() наоборот, берет одноканальные изображения и объединяет их для создания многоканальной картинки. Используя эти функции, получим 5 изображений с замененными местами каналами

```python
r,g,b = cv2.split(img)
img2 = cv2.merge((b,g,r))

r,g,b = cv2.split(img)
img3 = cv2.merge((g,b,r))

r,g,b = cv2.split(img)
img4 = cv2.merge((g,r,b))

r,g,b = cv2.split(img)
img5 = cv2.merge((b,r,g))

r,g,b = cv2.split(img)
img6 = cv2.merge((r,b,g))
```

Теперь осталось отобразить оригинальное и полученные изображения, использую библиотеку matplotlib

```python
plt.subplot(231)
plt.imshow(img)
plt.axis("off")

plt.subplot(232)
plt.imshow(img2)
plt.axis("off")

plt.subplot(233)
plt.imshow(img3)
plt.axis("off")

plt.subplot(234)
plt.imshow(img4)
plt.axis("off")

plt.subplot(235)
plt.imshow(img5)
plt.axis("off")

plt.subplot(236)
plt.imshow(img6)
plt.axis("off")

plt.show()
```
